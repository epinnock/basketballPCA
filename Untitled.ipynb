```python
import numpy as np

player_stats=np.array(player_df[['PTS','REB','AST','STL','BLK','TOV','FTA',To'FG3M']])
print player_stats
```

Import sklearn and perform PCA

```python
from sklearn.decomposition import PCA 

player_pca=PCA(n_components=3) #we are crushing things down to 3 dimensions
skt=player_pca.fit_transform(player_stats)
```

```python
import math
def distance(player1,player2):
    player1=player_and_loc_df.loc[player_and_loc_df['PLAYER_NAME']==player1].values.tolist()[0]
    player2=player_and_loc_df.loc[player_and_loc_df['PLAYER_NAME']==player2].values.tolist()[0]
    ret=math.sqrt(math.pow((player1[1]-player2[1]),2)+math.pow((player1[2]-player2[2]),2)+math.pow((player1[3]-player2[3]),2))
    return ret
print distance('Chris Paul','Ty Lawson')
print distance('Chris Paul','John Wall')
print distance('Dirk Nowitzki','Carmelo Anthony')
print distance('James Harden','Damian Lillard')
print distance('Kawhi Leonard','Tony Allen')
print distance('Kawhi Leonard','Gordon Hayward')
print distance('Russell Westbrook','James Harden')
```

```python
def find_n_closest(player1,n):
    names=player_and_loc_df['PLAYER_NAME'].values.tolist()
    near=[(name,distance(player1,name)) for name in names]
    near=sorted(near,key=lambda k: k[1])
    ret=near[:n]
    return ret
    
print ('Lebron James',find_n_closest('LeBron James',9))
distance('Tyson Chandler','DeAndre Jordan')
```
